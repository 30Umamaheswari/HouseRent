```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house_train = pd.read_excel('House_Rent_Train.xlsx')
house_train.head(2)
```

```python
house_train.shape
```

```python
house_train = house_train.dropna(subset=['rent','amenities'])
```

```python
import json

# Extract amenities information
amenities_list = house_train['amenities']

# Initialize a dictionary to store extracted values
amenities_dict = {}

# Iterate through each row
for am in amenities_list:
    amen_dict = json.loads(am)

    # Ensure that all keys are present in each row
    for key in amen_dict.keys():
        if key not in amenities_dict:
            amenities_dict[key] = []

    # Append values to the dictionary, filling with None if key is missing
    for key in amenities_dict.keys():
        amenities_dict[key].append(amen_dict.get(key, False))

# Add new columns for each amenity to the existing DataFrame
for key, values in amenities_dict.items():
    house_train[key] = values

# Display the modified DataFram
house_train
```

```python
house_train = house_train.dropna(subset=['type','locality'])
```

```python
house_train.isna().sum()
```

```python
house_train.shape
```

```python
house_train['GYM'] = house_train['GYM'].astype(int)
house_train['LIFT'] = house_train['LIFT'].astype(int)
house_train['INTERNET'] = house_train['INTERNET'].astype(int)
house_train['AC'] = house_train['AC'].astype(int)
house_train['CLUB'] = house_train['CLUB'].astype(int)
house_train['INTERCOM'] = house_train['INTERCOM'].astype(int)
house_train['POOL'] = house_train['POOL'].astype(int)
house_train['CPA'] = house_train['CPA'].astype(int)
house_train['FS'] = house_train['FS'].astype(int)
house_train['SERVANT'] = house_train['SERVANT'].astype(int)
house_train['SECURITY'] = house_train['SECURITY'].astype(int)
house_train['SC'] = house_train['SC'].astype(int)
house_train['GP'] = house_train['GP'].astype(int)
house_train['PARK'] = house_train['PARK'].astype(int)
house_train['RWH'] = house_train['RWH'].astype(int)
house_train['STP'] = house_train['STP'].astype(int)
house_train['HK'] = house_train['HK'].astype(int)
house_train['PB'] = house_train['PB'].astype(int)
house_train['VP'] = house_train['VP'].astype(int)
```

```python
house_train['bathroom'] = house_train['bathroom'].astype(int)
house_train['cup_board'] = house_train['cup_board'].astype(int)
house_train['floor'] = house_train['floor'].astype(int)
house_train['total_floor'] = house_train['total_floor'].astype(int)
```

```python
house_train['property_size'] = house_train['property_size'].astype(int)
house_train['property_age'] = house_train['property_age'].astype(int)
house_train['balconies'] = house_train['balconies'].astype(int)
house_train['rent'] = house_train['rent'].astype(int)
house_train['negotiable'] = house_train['negotiable'].astype(int)

house_train['latitude'] = house_train['latitude'].astype(float)
house_train['longitude'] = house_train['longitude'].astype(float)

house_train['locality'] = house_train['locality'].astype(str)
house_train['type'] = house_train['type'].astype(str)
house_train['id'] = house_train['id'].astype(str)
house_train['lease_type'] = house_train['lease_type'].astype(str)
house_train['furnishing'] = house_train['furnishing'].astype(str)
house_train['parking'] = house_train['parking'].astype(str)
house_train['facing'] = house_train['facing'].astype(str)
house_train['water_supply'] = house_train['water_supply'].astype(str)
house_train['building_type'] = house_train['building_type'].astype(str)
```

```python
remove_columns = ['gym', 'lift', 'swimming_pool','amenities']
house_train = house_train.drop(columns = remove_columns)
```

```python
house_train = house_train.rename(columns={'id':'ID','type':'House Type','locality':'Locality','activation_date':'Activation Date','latitude':'Latitude','longitude':'Longitude',
                            'lease_type':'Lease Type','furnishing':'Furnishing','parking':'Parking','property_size':'Property Size','property_age':'Property Age',
                            'bathroom':'No of Bathroom','facing':'House Facing','cup_board':'No of Cup Board','floor':'Floor No Floor','water_supply':'Water Supply',
                            'building_type':'Building Type','balconies':'Balconies','rent':'Rent','negotiable':'Negotiable','LIFT':'Lift', 'GYM':'Gym', 'INTERNET':'Internet',
                            'AC':'Air Conditioning','CLUB':'Club', 'INTERCOM':'Intercom', 'POOL':'Swimming Pool', 'CPA':'Car Parking', 'FS':' Feng Shui/Vaastu', 'SERVANT':'Servent',
                            'SECURITY':'Security', 'SC':'Shopping Center','GP':'Gas Pipeline', 'PARK':'Park', 'RWH':'Rain Water Harvesting', 'STP':'Sewage Treatment Plant',
                            'HK':'House Keeping', 'PB':'Power Backup', 'VP':'Visitor Parking'})
```
